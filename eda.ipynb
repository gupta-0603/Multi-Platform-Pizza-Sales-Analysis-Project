### Daily trend - Total orders

```python
df=pd.read_csv(r"C:\Users\Aditya Kumar\OneDrive\Desktop\pizza_sales.csv")
df['order_date']= pd.to_datetime(df['order_date'],dayfirst=True) #converts the order_date column to datetime format
df['dayname']= df['order_date'].dt.day_name() #extracts the day name from the order_date column and creates a new column called dayname
weekday_order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'] #defines the order of days

df['dayname'] = pd.Categorical(df['dayname'], categories=weekday_order, ordered=True)
orders_by_day = df.groupby('dayname', observed=False)['order_id'].nunique().reindex(weekday_order, fill_value=0) #groups the data by dayname and counts the number of unique orders for each day, then reindexes the result to match the order of days defined earlier
ax = orders_by_day.plot(kind='bar', color='green', figsize=(10,6), edgecolor='black')
plt.title('Total Orders by Day of the Week') #sets the title of the plot
plt.xlabel('Day of  Week') #sets the x-axis label
plt.ylabel('Number of Orders') #sets the y-axis label   
plt.xticks(rotation=45)
for i, (key, value) in enumerate(orders_by_day.items()):
    plt.text(i, value + 20, str(value), ha='center', va='bottom', fontsize=12, fontweight='bold') #adds the count of orders on top of each bar
plt.tight_layout() #adjusts the layout of the plot to prevent overlapping elements  
plt.show()

#displays the plot  
#   
```

### Daily Trend - Total Revenue

```python
df=pd.read_csv(r"C:\Users\Aditya Kumar\OneDrive\Desktop\pizza_sales.csv")
df['order_date']= pd.to_datetime(df['order_date'],dayfirst=True) #converts the order_date column to datetime format
df['dayname']= df['order_date'].dt.day_name() #extracts the day name from the order_date column and creates a new column called dayname
weekday_order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'] #defines the order of days

df['dayname'] = pd.Categorical(df['dayname'], categories=weekday_order, ordered=True)
orders_by_day = df.groupby('dayname', observed=False)['total_price'].sum() #groups the data by dayname and counts the number of unique orders for each day, then reindexes the result to match the order of days defined earlier
ax = orders_by_day.plot(kind='bar', color='purple', figsize=(10,6), edgecolor='black')
plt.title('Total Revenue by Day of the Week') #sets the title of the plot
plt.xlabel('Day of  Week') #sets the x-axis label
plt.ylabel('Total Revenue') #sets the y-axis label   
plt.xticks(rotation=45)
for i, (key, value) in enumerate(orders_by_day.items()):
    plt.text(i, value + 20, str(value), ha='center', va='bottom', fontsize=12, fontweight='bold') #adds the count of orders on top of each bar
plt.tight_layout() #adjusts the layout of the plot to prevent overlapping elements  
plt.show()

#displays the plot  
#   
```

### Hourly trend - Total Orders

```python
df=pd.read_csv(r"C:\Users\Aditya Kumar\OneDrive\Desktop\pizza_sales.csv")
df['order_time']= pd.to_datetime(df['order_time'],format='%H:%M:%S') #converts the order_date column to datetime format
df['order_hour']= df['order_time'].dt.hour #extracts the hour from the order_time column and creates a new column called order_hour
orders_by_hour = df.groupby('order_hour', observed=False)['order_id'].nunique() #groups the data by order_hour and counts the number of unique orders for each hour, then reindexes the result to match the order of hours defined earlier
ax = orders_by_hour.plot(kind='bar', color='red', figsize=(10,5), edgecolor='black')
plt.title('Total Orders by Hour of the Day') #sets the title of the plot
plt.xlabel('Hour of the Day') #sets the x-axis label
plt.ylabel('Total Orders') #sets the y-axis label   
plt.xticks(rotation=45)
for i, (key, value) in enumerate(orders_by_hour.items()):
    plt.text(i, value + 5, str(value), ha='center', va='bottom', fontsize=12, fontweight='bold') #adds the count of orders on top of each bar
plt.tight_layout() #adjusts the layout of the plot to prevent overlapping elements  
plt.show()

#displays the plot  
#   
```

### Monthly trend -Total Orders

```python
# monthly trend total orders by area line chart
df=pd.read_csv(r"C:\Users\Aditya Kumar\OneDrive\Desktop\pizza_sales.csv")
df['order_date']= pd.to_datetime(df['order_date'],dayfirst=True) #converts the order_date column to datetime format
df['month_name']= df['order_date'].dt.month_name() #extracts the month name from the order_date column and creates a new column called month_name
month_order=['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'] 
df['month_order'] = pd.Categorical(df['month_name'], categories=month_order, ordered=True)
orders_by_month = df.groupby('month_order', observed=False)['order_id'].nunique() #groups the data by month_name and counts the number of unique orders for each month, then reindexes the result to match the order of months defined earlier
plt.figure(figsize=(10,6)) #sets the size of the plot
plt.fill_between(orders_by_month.index, orders_by_month.values, color='skyblue', alpha=0.4) #creates a filled area plot
plt.plot(orders_by_month.index, orders_by_month.values, color='black', linewidth=2, marker='o') #creates a line plot
plt.title('Total Orders by Month') #sets the title of the plot
plt.xlabel('Month') #sets the x-axis label
plt.ylabel('Number of Orders') #sets the y-axis label   
plt.xticks(rotation=45)
for i, (key, value) in enumerate(orders_by_month.items()):
    plt.text(i, value + 20, str(value), ha='center', va='bottom', fontsize=12, fontweight='bold') #adds the count of orders on top of each bar
plt.tight_layout() #adjusts the layout of the plot to prevent overlapping elements  
plt.show()

#displays the plot  
```

### $ of sales by pizza category

```python
# % of sales by pizza category by donut chart
category_sales = df.groupby('pizza_category')['total_price'].sum() #groups the data by pizza_category and sums the total_price for each category
category_pct= category_sales/category_sales.sum()*100 #calculates the percentage of sales for each pizza category
plt.figure(figsize=(7,7)) #sets the size of the plot
colors=plt.get_cmap('tab20').colors #sets the colors for the pie chart
plt.pie(category_pct, labels=category_pct.index, autopct='%1.1f%%', startangle=90, colors=colors, wedgeprops=dict(width=0.4)) #creates a donut chart
plt.title('Percentage of Sales by Pizza Category') #sets the title of the plot
plt.tight_layout() #adjusts the layout of the plot to prevent overlapping elements
plt.show()
```

## % of sales by pizza size and category

```python
sales_pivot = df.pivot_table(
    index='pizza_category',
    columns='pizza_size',
    values='total_price',
    aggfunc='sum',
    fill_value=0
) #creates a pivot table to calculate the total sales for each pizza category
sales_pct=sales_pivot/sales_pivot.sum().sum() * 100 #calculates the percentage of sales for each pizza category and size
plt.figure(figsize=(10,6)) #sets the size of the plot
sns.heatmap(sales_pct, annot=True, fmt='.1f', cmap='YlOrRd', linewidths=.5) #creates a heatmap to visualize the percentage of sales for each pizza category and size
plt.title('Percentage of Sales by Pizza Category and Size') #sets the title of the plot
plt.xlabel('Pizza Size') #sets the x-axis label
plt.ylabel('Pizza Category') #sets the y-axis label
plt.tight_layout() #adjusts the layout of the plot to prevent overlapping elements
plt.show() #displays the plot
```
